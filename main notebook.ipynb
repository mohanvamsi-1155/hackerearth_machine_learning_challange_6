```python
combined_train=pd.read_csv('combined_train.csv')
combined_test=pd.read_csv('combined_test.csv')
```

```python
combined_train.info()
```

```python
combined_train['has_repair_started'].mode()
```

```python
combined_train['has_repair_started'].fillna(0.0,inplace=True)
```

```python
combined_train.info()
```

```python
id=combined_train['building_id']
```

```python
combined_train.drop('building_id',axis=1,inplace=True)
```

```python
for i in range(len(id)):
    if(pd.isnull(combined_train['count_families'][i])):
        combined_train['count_families'][i]=1
        print(i,id[i])
        break
```

```python
combined_train.info()
```

```python
combined_test.info()
```

```python
from sklearn.preprocessing import LabelEncoder 
le=LabelEncoder()
label=le.fit_transform(combined_train['damage_grade'])
```

```python
label
```

```python
answer_label=pd.DataFrame(label,columns=['damage_grade'])
```

```python
combined_train.drop('damage_grade',axis=1,inplace=True)
combined_train=pd.concat([combined_train,answer_label],axis=1)
```

```python
combined_train.info()
```

```python
area=pd.get_dummies(combined_train['area_assesed'],drop_first=True)
land=pd.get_dummies(combined_train['land_surface_condition'],drop_first=True)
foundation=pd.get_dummies(combined_train['foundation_type'],drop_first=True)
foundation.columns=['a','b','c','d']
roof=pd.get_dummies(combined_train['roof_type'],drop_first=True)
roof.columns=['e','f']
ground=pd.get_dummies(combined_train['ground_floor_type'],drop_first=True)
other=pd.get_dummies(combined_train['other_floor_type'],drop_first=True)
pos=pd.get_dummies(combined_train['position'],drop_first=True)
plan=pd.get_dummies(combined_train['plan_configuration'],drop_first=True)
cond=pd.get_dummies(combined_train['condition_post_eq'],drop_first=True)
owner=pd.get_dummies(combined_train['legal_ownership_status'],drop_first=True)
combined_train=pd.concat([combined_train,area,land,foundation,roof,ground,other,pos,plan,cond,owner],axis=1)
combined_train.drop(['area_assesed','land_surface_condition','foundation_type','roof_type','ground_floor_type','other_floor_type','position','plan_configuration','condition_post_eq','legal_ownership_status'],axis=1,inplace=True)
```

```python
combined_train.drop(['E_shape','H_shape','L_shape','Multi-projected','Others','Rectangular','Square','T-shape','U-shape'],axis=1,inplace=True)
```

```python
combined_train.drop(['has_secondary_use_agriculture','has_secondary_use_institution','has_secondary_use_health_post','has_secondary_use_other'],axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(combined_train.drop('damage_grade',axis=1),combined_train['damage_grade'],test_size=0.01,random_state=53)
```

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost.sklearn import XGBClassifier
```

```python
rf=RandomForestClassifier(random_state=111)
rf.fit(X_train,y_train)
```

```python
print(rf.get_params)
```

```python
predicted=rf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,f1_score
```

```python
confusion_matrix(y_test,predicted)
```

```python
accuracy_score(y_test,predicted)
```
